```python
import dask.dataframe as dd
import pandas as pd
import os
import re
```

```python
directory = r'E:\Senior Project\dataset'

csv_files = [f for f in os.listdir(directory) if f.endswith('.csv')]

def custom_sort(filename):
    if filename == 'asteroid_last.csv':
        return float('inf') 
    match = re.search(r'\d+', filename)
    return int(match.group()) if match else 0

csv_files = sorted(csv_files, key=custom_sort)

print(csv_files)
```

```python
df = pd.read_csv(os.path.join(directory, csv_files[0]))

for i in range(1, 7):
    pdf = pd.read_csv(os.path.join(directory, csv_files[i]))
    df = pd.merge(left=df, right=pdf, on='jd', how='inner')
    print(f"Complete {i}")
```
