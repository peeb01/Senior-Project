```python
import tensorflow as tf
import os

directory = 'E:\Senior Project\Coding File\Asteroids Pos'
csv_files = ['asteroid_from_0_to_600.csv', 'asteroid_from_600_to_1200.csv', 'asteroid_from_1200_to_1800.csv']

file_path = 'float32_asteroid_from_0_to_600.csv'

with open(file_path, 'r') as f:
    header = f.readline().strip().split(',')

columns_to_read = header[1:]

def cast_to_float32(features):
    return {key: tf.cast(value, tf.float32) for key, value in features.items()}

dataset = tf.data.experimental.make_csv_dataset(
    file_path,
    batch_size=64,    
    na_value='',      
    num_epochs=1,     
    ignore_errors=True,
    select_columns=columns_to_read
)

dataset = dataset.map(cast_to_float32)
```

```python
all_batches = []

for batch in dataset:
    batch_array = tf.concat(list(batch.values()), axis=-1)
    all_batches.append(batch_array)

tf_array = tf.concat(all_batches, axis=0)

print(tf_array)
```
